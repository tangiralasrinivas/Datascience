## Area/ Country Based Analysis

```python
import pandas as pd
import seaborn as sns
```

```python
hdl_data = pd.read_excel("HDL_Data.xls")
```

```python
hdl_data.columns
```

```python
hdl_area_data_hd = hdl_data[["HDcount","pop_2010","ownhome_2010","income_2010","density_2010",\
                              "pct_U18_2010","pctblack_2010","pctcollege_2010","pctwhite_2010",\
                             "pop_2000","ownhome_2000","income_2000","density_2000",\
                              "pct_U18_2000","pctblack_2000","pctcollege_2000","pctwhite_2000"]]
hdl_area_data_l = hdl_data[["Lcount","pop_2010","ownhome_2010","income_2010","density_2010",\
                              "pct_U18_2010","pctblack_2010","pctcollege_2010","pctwhite_2010",\
                             "pop_2000","ownhome_2000","income_2000","density_2000",\
                              "pct_U18_2000","pctblack_2000","pctcollege_2000","pctwhite_2000"]]

hdl_area_data_hd["Flag"] = "Home Depot"
hdl_area_data_l["Flag"] = "Lowes"
```

```python
hdl_area_data_hd = hdl_area_data_hd[hdl_area_data_hd["HDcount"]>0]
hdl_area_data_hd = hdl_area_data_hd[hdl_area_data_hd["density_2010"]<7600]
hdl_area_data_hd = hdl_area_data_hd[hdl_area_data_hd["density_2000"]<7600]
hdl_area_data_l = hdl_area_data_l[hdl_area_data_l["Lcount"]>0]
hdl_area_data_l = hdl_area_data_l[hdl_area_data_l["density_2010"]< 7600]
hdl_area_data_l = hdl_area_data_l[hdl_area_data_l["density_2000"]<7600]
```

```python
hdl_area_data_hd["density_2010"].max()
```

```python
hdl_area_data_hd = hdl_area_data_hd.loc[hdl_area_data_hd.index.\
                                       repeat(hdl_area_data_hd["HDcount"])].reset_index(drop=True)\
                               [["pop_2010","ownhome_2010","income_2010","density_2010",\
                              "pct_U18_2010","pctblack_2010","pctcollege_2010","pctwhite_2010",\
                             "pop_2000","ownhome_2000","income_2000","density_2000",\
                              "pct_U18_2000","pctblack_2000","pctcollege_2000","pctwhite_2000","Flag"]]
hdl_area_data_l = hdl_area_data_l.loc[hdl_area_data_l.index.\
                                       repeat(hdl_area_data_l["Lcount"])].reset_index(drop=True)\
                               [["pop_2010","ownhome_2010","income_2010","density_2010",\
                              "pct_U18_2010","pctblack_2010","pctcollege_2010","pctwhite_2010",\
                             "pop_2000","ownhome_2000","income_2000","density_2000",\
                              "pct_U18_2000","pctblack_2000","pctcollege_2000","pctwhite_2000","Flag"]]
hdl_area = pd.concat([hdl_area_data_hd,hdl_area_data_l])
```

```python
sns.set(style="ticks")
plt.figure(figsize=(15,15))
plt.subplot(2,2,1)
sns.boxplot(x="Flag",y="income_2010",data=hdl_area,palette="GnBu_d")
plt.subplot(2,2,2)
sns.boxplot(x="Flag",y="ownhome_2010",data=hdl_area,palette="GnBu_d")
plt.subplot(2,2,3)
sns.boxplot(x="Flag",y="pctblack_2010",data=hdl_area,palette="GnBu_d")
plt.subplot(2,2,4)
sns.boxplot(x="Flag",y="pct_U18_2010",data=hdl_area,palette="GnBu_d")
```

```python
sns.set(style="ticks")
plt.figure(figsize=(15,15))
plt.subplot(2,2,1)
sns.boxplot(x="Flag",y="density_2010",data=hdl_area,palette="GnBu_d")
plt.subplot(2,2,2)
sns.boxplot(x="Flag",y="pop_2010",data=hdl_area,palette="GnBu_d")
plt.subplot(2,2,3)
sns.boxplot(x="Flag",y="pctcollege_2010",data=hdl_area,palette="GnBu_d")
plt.subplot(2,2,4)
sns.boxplot(x="Flag",y="pctwhite_2010",data=hdl_area,palette="GnBu_d")
```
